```python
import csv

def ler_horas_csv(arquivo):

  try:
    with open(arquivo, newline='', encoding='utf-8') as f:
      leitor = csv.DictReader(f)
      for linha in leitor:
        yield linha
  except FileNotFoundError:
    print(f"Erro: O arquivo '{arquivo}' não foi encontrado  !!!")
    return

def calcular_horas_por_projeto(arquivo):
  projetos = {}
  desenvolvedores = set()

  try:
    for linha in ler_horas_csv(arquivo):
      if linha is None:
        return

      projeto = linha["Projeto"]
      horas = int(linha["Horas"])
      dev = linha["Desenvolvedor"]

    # Soamndo horas

      if projeto in projetos:
        projetos[projeto] += horas

      else:
        projetos[projeto] = horas

      desenvolvedores.add(dev)
  except ValueError:
    print(f"AVISO : erro de linha de formatação")

  return projetos , desenvolvedores


# Executando

arquivo = "horas.csv"

projetos , desenvolvedores = calcular_horas_por_projeto(arquivo)


if projetos:
  print("-"*20)
  print("Total de horas por projetos:")
  for projeto, horas in projetos.items():
    print(f"- {projeto}: {horas} horas")

  print("-"*20)
  print("Desenvolvedores únicos")
  print(" ,".join(desenvolvedores))


```
